```python
## 假如我們不想把資料載到自己的電腦裡?
# 把連結填入
target_url = 'https://raw.githubusercontent.com/vashineyu/slides_and_others/master/tutorial/examples/imagenet_urls_examples.txt'
```

```python
import requests
response = requests.get(target_url)
data = response.text

# 用 request 傳送回來的資料不會認得斷行符號
print(len(data))
data[0:100]
```

```python
# 找到換行符號，用該符號做字串分割後，把它拿掉
split_tag = '\n'

data = data.split(split_tag)
print(len(data))
data[0]
```

## 將 txt 轉成 pandas dataframe

```python
import pandas as pd

arrange_data =[]
for t in data:
  arrange_data.append(t.split('\t'))

df = pd.DataFrame(arrange_data, columns = ["ImagName", "ImagUrl"])
df.head()
```

## 讀取圖片，請讀取上面 data frame 中的前 5 張圖片

```python
from PIL import Image
from IPython.display import display #mac 顯示圖片使用
from io import BytesIO
import numpy as np
import matplotlib.pyplot as plt

# 請用 df.loc[...] 得到第一筆資料的連結
first_link = df.loc[0]['ImagUrl'];

response = requests.get(first_link)
img = Image.open(BytesIO(response.content))

#im_array = np.asarray(img)
display(img)
#plt.imshow(im_array)
#plt.show()

#print(df[0:5]);
# Convert img to numpy array

#img.imshow()
#img.show()
```

```python
def img2arr_fromURLs(url_list, resize = False):
    """
    請完成這個 Function
    Args
        - url_list: list of URLs
        - resize: bool
    Return
        - list of array
    """
    img_list=[]
    
    for i in range(len(url_list)):
        try:
            response = requests.get(url_list.loc[i]['ImagUrl'])
            img = Image.open(BytesIO(response.content))
            img_list.append(img)
        except:
            print("except")
        
    return img_list
```

```python
result = img2arr_fromURLs(df[0:5])
print("Total images that we got: %i " % len(result)) # 如果不等於 5, 代表有些連結失效囉

for im_get in result:
    display(im_get)
    #plt.imshow(im_get)
    #plt.show()
```
